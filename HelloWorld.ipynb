```python
import openai
```

```python
# openai API 키 인증
openai.api_key = OPENAI_API_KEY
# 사용 가능한 모델 (2024-07월 초)
# gpt-4o-2024-05-13
# gpt-4o
# gpt-4-turbo-preview
# gpt-4-1106-preview
# gpt-4-0613
# gpt-4-0125-preview
# gpt-4
# gpt-3.5-turbo-16k
# gpt-3.5-turbo-1106
# gpt-3.5-burbo-0125
# gpt-3.5-turbo
model = "gpt-3.5-turbo-0125"
```

```python
# 질문 작성하기
query = "프랑스의 수도를 알려줘"

# 메시지 설정하기
messages = [{
    "role":"user",
    "content":query
}]
```

```python
# ChatGPT API를 이용한 ChatGPT 사용하기
response = openai.ChatCompletion.create(
    model=model,
    messages=messages
)
answer = response.choices[0].message.content
print(answer)
```
